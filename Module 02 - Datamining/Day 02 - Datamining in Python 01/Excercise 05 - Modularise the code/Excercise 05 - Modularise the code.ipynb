#Modularise the code

Sometimes your code grows too large and you like to divide it into various - smaller - segments. Additionally, this approach offers the benefit of being able to utilize the same code across multiple projects.

Lastly, you may want to maintain your passwords or log in details separately from the main code. This method enables you to share your code with others while still being able to sleep at night.

##Save the variables outside your main code

This represents the simplest choice. Generate a text file containing the credentials alongside your code, then change the file extension from *.txt to *.py. Subsequently, you can include the file in your primary script.

The exercise folder contains a file called `"credits.py"`

First, we connect the drive to gain access. When running the code, you'll be prompted to grant Google Colab permission to mount the drive.

```python
from google.colab import drive
drive.mount('/content/drive')
```

Next, you append the directory containing the credits as a system path. Once this is completed, you can import the contents of the file.

```python
import sys
sys.path.append('/content/drive/MyDrive/Colab Notebooks/RC15 23 Excercises/Excercise 05 - Modularise the code')

import credits

print (credits.a)
print (credits.key)
```

## Save functions outside your code

The exercise folder also includes some basic functions that we can integrate into our main script. Assuming you've connected the drive and executed the sys.path.append command from the previous example, all that's left is to perform the following:

```python
import functions
```

```python
print (functions.sayHello())
print (functions.addTwo(3))
```

To avoid writing `functions` or `credits` all the time, utilize a shortcut:

```python
import credits as cr
import functions as fn
```

```python
print (cr.a)
print (fn.addTwo(5))
```
